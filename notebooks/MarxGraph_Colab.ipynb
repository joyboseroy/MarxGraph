# MarxGraph pipeline (Colab)

**Note on hardware:** this notebook calls a hosted LLM API (Groq or Anthropic) for every extraction step. Nothing here runs a model locally, so a GPU runtime does not speed anything up — the bottleneck is network-bound (waiting on API responses), which is fixed by running requests **concurrently**, not by hardware. Use **Runtime > Change runtime type > CPU** to avoid tying up a GPU allocation you don't need.

This notebook mounts Google Drive so `data/raw` (EPUBs) and `data/parquet` (extracted tables) persist across sessions — Colab's local disk is wiped when the runtime recycles, and this pipeline is designed to be resumable across many short sessions.

## 4. Set your API key
Choose ONE backend. Groq is much cheaper for a corpus this size; Anthropic gives higher-quality extraction. The key is entered via a hidden prompt (`getpass`) so it never gets saved into the notebook file itself.

```python
import os
from getpass import getpass

BACKEND = 'groq'  # or 'anthropic'
os.environ['MARXGRAPH_BACKEND'] = BACKEND

if BACKEND == 'groq':
    os.environ['GROQ_API_KEY'] = getpass('Groq API key: ')
else:
    os.environ['ANTHROPIC_API_KEY'] = getpass('Anthropic API key: ')
print('Backend set to', BACKEND)
```
